#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
median_steps_month = fitbit.groupby('Months_encoded').agg({'Steps':'median'}).reset_index()

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
median_work_weekend = fitbit.groupby('Work_or_Weekend').agg({'Steps':'median'}).reset_index()

_ , (chart_1, chart_2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

chart_1.plot(median_steps_month['Months_encoded'], median_steps_month['Steps'], '-x')
chart_2.bar(median_work_weekend['Work_or_Weekend'], median_work_weekend['Steps'])
plt.savefig('chart.png')
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
act_minutes = fitbit.groupby(['Minutes Lightly Active','Minutes Very Active','Minutes Sedentary']).agg({'Steps':'median'}).reset_index()
act_list = ['Minutes Lightly Active','Minutes Very Active','Minutes Sedentary']


for act in act_list:
    title = act+' vs Steps'
    plt.title(title)
    plt.scatter(act_minutes[act], act_minutes['Steps'])
    plt.show()
```
